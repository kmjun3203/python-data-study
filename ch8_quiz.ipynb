```python
# 필요한 모듈 가져오기
import numpy as np
# 맷플롯립은 여러 모듈로 구성됨
# 그 중 그래프를 제공하는 모듈 꺼내기
import matplotlib.pyplot as plt
# 그래프를 그릴때 한글이 깨지는 현상이 발생
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

```python
# 시본도 한글 폰트 설정 필요
import pandas as pd
import seaborn as sns
sns.set_theme(font = 'Malgun Gothic')
```

```python
# 일주일간의 평균 기온 데이터
df = pd.DataFrame({'요일': ['월','화','수','목','금','토','일'], '기온': [12, 14, 13, 15, 17, 18, 16]})

sns.lineplot(data = df, x = '요일', y = '기온')
plt.title('일주일 평균 기온')
```

```python
# 과일별 하루 판매량 데이터
df = pd.DataFrame({'과일': ['사과','바나나','포도','딸기'],
                   '판매량': [120, 80, 150, 100]})
sns.barplot(data = df, x = '과일', y = '판매량')
plt.title('과일별 판매량')
```

```python
# 학생들의 공부 시간과 시험 점수 데이터
df = pd.DataFrame({'공부시간': [2, 4, 6, 3, 5, 7, 8],
                   '점수': [55, 65, 75, 60, 70, 85, 90]})
sns.scatterplot(data = df, x = '공부시간', y = '점수')
plt.title('공부시간')
```

```python
# 50명의 하루 걸음 수 데이터
arr = np.random.randint(3000, 10001, 50)
df = pd.DataFrame({'걸음 수': arr})
sns.histplot(data = df, x = '걸음 수', bins = 7)
plt.title('하루 걸음 수 분포')
plt.ylabel('사람 수')
```

```python
df = pd.DataFrame({'요일' : ['월','화','수','목','금','토','일'],
                   '기온' : [12,14,13,15,17,18,16]})
fig, axs = plt.subplots(1, 2, figsize = (10,3))
sns.lineplot(data = df, x = '요일', y = '기온', ax = axs[0])
axs[0].set_title('요일별 기온(선)')
sns.scatterplot(data = df, x = '요일', y = '기온', ax = axs[1])
axs[1].set_title('요일별 기온(산점도)')
plt.tight_layout()
```

```python
# 과일별 하루 판매량 데이터
df1 = pd.DataFrame({'과일': ['사과','바나나','포도','딸기'],
                   '판매량': [120, 80, 150, 100]})
# 학생 점수 데이터
# 학생들의 공부 시간과 시험 점수 데이터
df2 = pd.DataFrame({'공부시간': [2, 4, 6, 3, 5, 7, 8],
                   '점수': [55, 65, 75, 60, 70, 85, 90]})
```

```python
fig, axs = plt.subplots(1, 3, figsize = (9, 3))
sns.barplot(data = df1, x = '과일', y = '판매량', ax = axs[0])
axs[0].set_title('과일 판매량(막대)')
sns.lineplot(data = df1, x = '과일', y = '판매량', ax = axs[1])
axs[1].set_title('과일 판매량(선)')
sns.scatterplot(data = df1, x = '과일', y = '판매량', ax = axs[2])
axs[2].set_title('과일 판매량(산점도)')

plt.tight_layout()
```

```python
fig, axs = plt.subplots(3, 1, figsize = (3, 9))
sns.scatterplot(data = df2, x = '공부시간', y = '점수', ax = axs[0])
axs[0].set_title('공부시간-점수(산점도)')
sns.barplot(data = df2, x = '공부시간', y = '점수', ax = axs[1])
axs[1].set_title('공부시간-점수(막대)')
sns.lineplot(data = df2, x = '공부시간', y = '점수', ax = axs[2])
axs[2].set_title('공부시간-점수(선)')
plt.tight_layout()
```

```python
df3 = pd.DataFrame({'이름' : ['영희','철수','민수','지연'],
                   '점수' : [85, 90, 80, 100]})
```

```python
fig, axs = plt.subplots(2, 2, figsize = (10,10))
sns.barplot(data = df1, x = '과일', y = '판매량', ax = axs[0,0])
axs[0,0].set_title('과일 판매량(막대)')
sns.lineplot(data = df1, x = '과일', y = '판매량', ax = axs[0,1])
axs[0,1].set_title('과일 판매량(선)')

sns.barplot(data = df3, x = '이름', y = '점수', ax = axs[1,0])
axs[1,0].set_title('학생별 점수(막대)')
sns.scatterplot(data = df3, x = '이름', y = '점수', ax = axs[1,1])
axs[1,1].set_title('학생별 점수(산점도)')
plt.tight_layout()
```

```python
df = pd.DataFrame({'구' : ['서구','서구','연수구','연수구','남동구','남동구'],
                   '인구(만명)' : [45,38,35,28,55,50],
                   '동' : ['연희동','가좌동','송도동','옥련동','구월동','논현동']})
df
```

```python
fig, axs = plt.subplots(1,2, figsize = (10,3))
sns.barplot(data = df, x = '동', y = '인구(만명)', palette = 'pastel', hue = '구', ax = axs[0])
axs[0].set_title('인천 구.동별 인구 현황(막대)', fontsize = 10, fontweight = 'bold')
axs[0].text('구월동', 55, '55만명', color = 'red', ha = 'center', va = 'bottom')
sns.lineplot(data = df, x = '동', y = '인구(만명)', linestyle = ':', color = 'blue', marker = 'o',ax = axs[1])
axs[1].set_title('인천 구.동별 인구 현황(선)', fontsize = 10, fontweight = 'bold')
plt.tight_layout
```

```python
df = pd.DataFrame({'카테고리' : ['식품','식품','의류','의류','전자제품','전자제품'],
                   '판매량' : [120,150,80,95,200,220],
                   '상품명' : ['라면','과자','셔츠','신발','노트북','스마트폰']})
df
```
