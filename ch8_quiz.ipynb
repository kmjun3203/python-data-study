```python
# 필요한 모듈 가져오기
import numpy as np
# 맷플롯립은 여러 모듈로 구성됨
# 그 중 그래프를 제공하는 모듈 꺼내기
import matplotlib.pyplot as plt
# 그래프를 그릴때 한글이 깨지는 현상이 발생
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

```python
a = np.arange(10,21)
plt.plot(a)
```

```python
a = ['사과','바나나','포도','오렌지']
b = [50, 80, 40, 70]
plt.bar(a, b)
```

```python
x = [1, 2, 3, 4, 5]
y = [40, 55, 65, 80, 85]
plt.scatter(x,y)
plt.title('산점도 그래프')
plt.xlabel('공부 시간')
plt.ylabel('시험 점수')
```

```python
a = np.random.randint(0, 11, 100)
plt.hist(a, bins = 10)
plt.title('유튜브 시청 시간 분포')
plt.xlabel('시청 시간(시간)')
plt.ylabel('사람 수')
```

```python
x = ['사과', '바나나', '포도', '오렌지', '복숭아']
y = [50, 80, 40, 70, 60]
figs, axs = plt.subplots(1,2,figsize = (10,3))
axs[0].plot(y)
axs[0].set_title('과일 판매량 (선 그래프)')
axs[1].bar(x,y)
plt.title('과일 판매량 (막대그래프)')
```

```python
x = ['월', '화', '수', '목', '금']
y = [2, 3, 4, 2, 5]
figs, axs = plt.subplots(1,2,figsize = (10,3))
axs[0].plot(y)
axs[0].set_title('요일별 유튜브 시청 시간 (선 그래프)')
axs[1].bar(x,y)
axs[1].set_title('요일별 유튜브 시청 시간 (막대그래프)')
```

```python
x = ['월', '화', '수', '목', '금']    
y = [1, 0.5, 2, 1.5, 2.5]
figs, axs = plt.subplots(1,2,figsize = (10,3))
axs[0].bar(x,y)
axs[0].set_title('요일별 독서 시간 (막대그래프)')
axs[1].scatter(x,y)
axs[1].set_title('요일별 독서 시간 (산점도)')
```

```python
x = ['서울', '부산', '대구', '광주', '제주']
y = [15.5, 17.8, 16.2, 15.0, 18.3]
figs, axs = plt.subplots(1,2,figsize = (10,3))
axs[0].plot(y)
axs[0].set_title('도시별 평균 기온 (선 그래프)')
axs[1].scatter(x,y)
axs[1].set_title('도시별 평균 기온 (산점도)')
```
