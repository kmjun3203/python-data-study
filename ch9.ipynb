### 인천 초중고 학교 지도 그리기

공공데이터 포털에서 전국 하굑 데이터를 가지고 온 후
필요한 데이터만 선택해서 지도 위에 그린다.

folum: 지도 기능을 제공하는 모듈

```python
# 모듈 불러오기
import pandas as pd
import folium
```

csv파일 불러오기

```python
# csv 파일을 읽어서 데이터 프레임으로 저장
data = pd.read_csv('전국초중등학교위치표준데이터.csv', encoding='euc-kr')
data
```

```python
# 여러 컬럼을 선택
df = data[['학교명', '시도교육청명', '학교급구분', '위도', '경도']]
df
# 인천에 있는 학교만 찾기
school = df[df['시도교육청명'] == '인천광역시교육청']
school
```

지도 만들기

```python
# 지도 객체 생성
# Map(지도 가운데 중심, 줌 레벨)
# 줌레벨이 높을수록 지도가 축소된다.
map = folium.Map(location = [37.456318,126.704973], zoom_start = 10)
map
```

지도 위에 마커 추가하기(학교 위치)
- 초중고에 따라 마커 색 다르게
- 초등학교orange
- 중학교green
- 고등학교blue

```python
# 데이터프레임 인덱스 확인
school.index
school.columns
# 데이터프레임 전체 데이터 확인
for i in school.index:
    name = school.loc[i, '학교명']
    typ = school.loc[i, '학교급구분']
    lat = school.loc[i, '위도']
    lon = school.loc[i, '경도']
    # 학교별 마커 만들기
    # 마커의 위치(경도 위도)

    # 조건문으로 마커의 색 결정하기
    # 학교급구분에 따라 색 결정
    if typ == '초등학교':
        colors = 'orange'
    elif typ == '중학교':
        colors = 'green'
    elif typ == '고등학교':
        colors = 'blue'
    # tooltip: 마우스를 올렸을 때만 정보가 표시됨
    # popup: 클릭하면 정보가 계속 표시됨
    folium.Marker([lat,lon], popup = name, tooltip = name, icon = folium.Icon(color = colors)).add_to(map)
map
```
