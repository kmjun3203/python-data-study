```python
import pandas as pd
import folium
```

```python
data = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20241028.csv', encoding='euc-kr')
data
```

```python
result1 = data[data['도시명'] == '인천광역시']
incheon = result1[result1['정류장명'].str.contains('마을회관')]
incheon
```

```python
result2 = data[data['도시명'] == '서울특별시']
seoul = result2[result2['정류장명'].str.contains('상가')]
seoul
```

```python
total = pd.concat([incheon, seoul], ignore_index= True)
total
```

```python
map1 = folium.Map(location = [37.496299,126.869525], zoom_start = 10)
map1
```

```python
total.columns
total.index
```

```python
for i in total.index:
    name = total.loc[i, '정류장명']
    typ = total.loc[i, '도시명']
    lat = total.loc[i, '위도']
    lon = total.loc[i, '경도']
    if typ == '인천광역시':
        colors = 'red'
    elif typ == '서울특별시':
        colors = 'blue'
    folium.Marker([lat,lon], popup = name, tooltip = name, icon = folium.Icon(color = colors)).add_to(map1)
```

```python
data = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20241028.csv', encoding='euc-kr')
df = data[['정류장명','위도','경도','도시명']]
df
```

```python
seoul1 = df[df['도시명'] == '서울특별시']
a = seoul1[seoul1['정류장명'].str.contains('상가')]
incheon1 = df[df['도시명'] == '인천광역시']
b = incheon1[incheon1['정류장명'].str.contains('행정복지센터')]
a
```

```python
result1 = pd.concat([a,b], ignore_index=True)
result1
```

```python
map = folium.Map(location = [37.496299,126.869525], zoom_start = 10)
map
```

```python
for i in result1.index:
    name = result1.loc[i, '정류장명']
    typ = result1.loc[i, '도시명']
    lat = result1.loc[i, '위도']
    lon = result1.loc[i, '경도']
    if typ == '서울특별시':
        colors = 'blue'
    elif typ == '인천광역시':
        colors = 'red'
    folium.Marker([lat,lon], popup=name, tooltip=name, icon=folium.Icon(color = colors)).add_to(map)
map
```
